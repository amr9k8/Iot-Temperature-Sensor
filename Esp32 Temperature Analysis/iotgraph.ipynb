```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
from mysql.connector import Error
import json
import urllib.request as urllib2
```

```python


url = r'http://tempsensor.atwebpages.com/Sensor_json.php?apikey=123123123'
response = urllib2.urlopen(url)
Sensor_table_data = json.load(response)
print(Sensor_table_data)

    
with open('temp_json.txt', 'w', encoding='utf-8') as f:
    json.dump(Sensor_table_data, f, ensure_ascii=False, indent=4)
```

```python
df = pd.read_json (r'temp_json.txt')
df.to_csv (r'New_temp.csv', index = None)
```
